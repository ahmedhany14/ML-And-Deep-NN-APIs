```python

class text_preprocessing():
    def __init__(
        self,
        lower=False,
        text_to_word=False,
        Join=False,
        remove_punct=False,
        remove_digits=False,
        remove_stopwords=False,
        steming=False,
        remove_whitespace=False,
        remove_html=False,
        remove_special=False,
        remove_urls=False,
        remove_emails=False,
        remove_emojis=False,
        remove_newlines=False,
        remove_hashtags=False,
    ):

        self.lower = lower
        self.text_to_word = text_to_word
        self.Join = Join
        self.steming = steming
        self.remove_punct = remove_punct
        self.remove_digits = remove_digits
        self.remove_stopwords = remove_stopwords
        self.remove_whitespace = remove_whitespace
        self.remove_html = remove_html
        self.remove_special = remove_special
        self.remove_urls = remove_urls
        self.remove_emails = remove_emails
        self.remove_emojis = remove_emojis
        self.remove_newlines = remove_newlines
        self.remove_hashtags = remove_hashtags


    # this function converts the input string into lowercase.
    def __Convert_to_lower(self, X=str):
        X = X.lower()

        return X

    # this function converts the input string into words.
    def __Convert_to_word(self, X=str):
        words = word_tokenize(X)
        return words

    def __Convert_to_sentence(self, X):
        sentence = " ".join(X)
        return sentence

    # this function removes punctuation from the input string.
    def __Remove_punctuation(self, X=str):
        X = re.sub(r"[^\w\s]", "", X)
        return X

    # this function removes digits from the input string.
    def __Remove_digits(self, X=str):
        X = re.sub(r"\d+", "", X)
        return X

    # this function removes stopwords from the input string.
    def __Remove_stopwords(self, X):
        fixed_words = []
        stop_words = set(stopwords.words("english"))
        for word in X:
            if word not in stop_words:
                fixed_words.append(word)
        return fixed_words

    def __Steming(self, X):
        fixed_words = []
        ps = PorterStemmer()

        for w in X:
            stem_w = ps.stem(w)
            fixed_words.append(stem_w)

        return fixed_words

    # this function removes leading and trailing whitespaces from the input string.
    def __Remove_whitespace(self, X):
        X = X.strip()
        return X

    # this function removes HTML tags from the input string.
    def __Remove_html(self, X=str):
        X = BeautifulSoup(X, "html.parser").get_text()
        return X

    # all the special characters like ! , : @ # $ % ^ & * ( ) have been removed from the input string.
    def __Remove_special(self, X):
        X = re.sub(r"[^a-zA-Z0-9\s]", "", X)
        
        print(X)
        print(type(X))
        return X

    # this function removes URLs from the input string.
    def __Remove_urls(self, X):
        X = re.sub(r"http\S+", "", X)
        return X

    # this function removes emails from the input string.
    def __Remove_emails(self, X):
        X = re.sub(r"\S*@\S*\s?", "", X)
        return X

    # this function removes emojis from the input string.
    def __Remove_emojis(self, X):
        X = X.encode("ascii", "ignore").decode("ascii")
        return X

    # this function removes newlines from the input string.
    def __Remove_newlines(self, X):
        X = re.sub(r"\n", "", X)
        return X

    # this function removes hashtags from the input string.
    def __Remove_hashtags(self, X):
        X = re.sub(r"#\w+", "", X)
        return X



    def transform(self, X):
        if self.lower:
            X = self.__Convert_to_lower(X)
        if self.text_to_word:
            X = self.__Convert_to_word(X)
        if self.remove_punct:
            X = self.__Remove_punctuation(X)
        if self.remove_digits:
            X = self.__Remove_digits(X)
        if self.remove_stopwords:
            X = self.__Remove_stopwords(X)
        if self.steming:
            X = self.__Steming(X)
        if self.remove_whitespace:
            X = self.__Remove_whitespace(X)
        if self.remove_html:
            X = self.__Remove_html(X)
        if self.remove_special:
            X = self.__Remove_special(X)
        if self.remove_urls:
            X = self.__Remove_urls(X)
        if self.remove_emails:
            X = self.__Remove_emails(X)
        if self.remove_emojis:
            X = self.__Remove_emojis(X)
        if self.remove_newlines:
            X = self.__Remove_newlines(X)
        if self.remove_hashtags:
            X = self.__Remove_hashtags(X)
        if self.Join:
            X = self.__Convert_to_sentence(X)
        return X
```
